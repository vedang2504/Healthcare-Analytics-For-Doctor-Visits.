# 10.Visualize the percentage of people getting govt health Insurance due to low income,due to old age and also the percentage of people having private health

```python
# % of people getting govt lesurance due to Low income
label = ['yes', 'no']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
X= [Y.shape[0], N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X, labels=label)
plt.title("% of people getting govt health Insurance due to low income ")
plt.show()
# % of of people having private Insurance
Y = df[df['private']=='yes']
N = df[df['private']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X, labels=label)
plt.title(" % of people having peivate health Insurance")
plt.show()
#% of people getting govt Insurance due to old age, disability ur veterem status
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
X = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(X,labels=label)
plt.title("% of people getting govt health Insurance due tasid age, disability veteran status")
plt.show()    
```

# 11.Plot a horizontal bar chart to analyze the reduced days of activity due to iilness based on Gender

```python
db=df.groupby('gender') ['reduced'].sum().to_frame().reset_index()
#Creating the bar chart
plt.barh(db['gender'], db['reduced'], color = ['cornflowerblue', 'lightseagreen'])
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('reduced activity')
plt.ylabel('gender')
#Show the plot
plt.show()
```
